## Elimina totales para no duplicar datos:

```python
#En el dataset aparecen filas con valores como "Total Imports", "Total Exports", "Distribution Losses" y "Final Consumption (Calculated)" en la columna "Balance".
#estas filas no representan un tipo específico de energía, sino que son totales o cálculos relacionados con el balance energético.
values_to_remove_balance = ["Total Imports", "Total Exports", "Distribution Losses", "Final Consumption (Calculated)", "Used for pumped storage"]

#También hay totales en la columna "Product" que duplican datos
values_to_remove_product = ["Total Combustible Fuels", "Total Renewables (Hydro, Geo, Solar, Wind, Other)"]

df_filtrado = df[
    ~df["Balance"].isin(values_to_remove_balance) &
    ~df["Product"].isin(values_to_remove_product)
]
```

## ✅ 3. Crear tabla agregada

```python
# Filtrar solo Renovable y No Renovable
df_final = df_filtrado[df_filtrado["Product"] != "No clasificado"]

tabla_resumen = df_final.groupby("Product")["Value"].sum().reset_index()
tabla_resumen["Value"] = tabla_resumen["Value"].astype(int)
# Agregar columna Unit
tabla_resumen["Unit"] = "GWh"
tabla_resumen = tabla_resumen[["Product", "Value", "Unit"]]
```

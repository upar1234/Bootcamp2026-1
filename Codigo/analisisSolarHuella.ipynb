# Análisis de distribución y normalidad 

```python
from scipy.stats import shapiro
```

```python
# Analisis de normalidad de los factores Energía [kWh/año] y Emisión CO2 [Ton/año] 
#Se realiza el análisis de normalidad utilizando el test de shapiro para las variables 'Energía [kWh/año]' y 'Emisiones CO2 [Ton/año]'. 

#Test de Shapiro-Wilk para 'Energía [kWh/año]'
stat_e, p_valor_e = shapiro(df['Energía [kWh/año]'])
print("=== Análisis de Normalidad: Energía [kWh/año] ===")
print(f"Estadístico de prueba: {stat_e:.4f}")
print(f"P-valor: {p_valor_e:.4e}")
if p_valor_e > 0.05:
    print("Conclusión: Los datos tienen distribución NORMAL (No se rechaza la hipótesis nula).\n")
else:
    print("Conclusión: Los datos tienen distribución NO NORMAL (Se rechaza la hipótesis nula).\n")
```

```python
#Test de Shapiro-Wilk para 'Emisiones CO2 [Ton/año]'
stat_c, p_valor_c = shapiro(df['Emisiones CO2 [Ton/año]'])
print("=== Análisis de Normalidad: Ahorro de Emisiones CO2 [Ton/año] ===")
print(f"Estadístico de prueba: {stat_c:.4f}")
print(f"P-valor: {p_valor_c:.4e}")
if p_valor_c > 0.05:
    print("Conclusión: Los datos tienen distribución NORMAL (No se rechaza la hipótesis nula).\n")
else:
    print("Conclusión: Los datos tienen distribución NO NORMAL (Se rechaza la hipótesis nula).\n")

```

# Análisis de correlación entre energía y emisiones

```python
# Calculo del coeficiente de Spearman
coef_spearman = df['Energía [kWh/año]'].corr(df['Emisiones CO2 [Ton/año]'], method='spearman')

print(f"Coeficiente de Correlación de Spearman (\u03c1): {coef_spearman:.4f}")

# lógica para la interpretación 
print("\n--- Interpretación ---")
if coef_spearman > 0:
    direccion = "Positiva (A mayor producción de energía solar, mayor ahorro de emisiones de CO2)"
else:
    direccion = "Negativa (A mayor producción de energía solar, menor ahorro de emisiones de CO2)"

if abs(coef_spearman) >= 0.8:
    fuerza = "Muy Fuerte"
elif abs(coef_spearman) >= 0.6:
    fuerza = "Fuerte"
elif abs(coef_spearman) >= 0.4:
    fuerza = "Moderada"
else:
    fuerza = "Débil"

print(f"Dirección: {direccion}")
print(f"Fuerza de la relación: {fuerza}")
```

# Construcción de modelo descriptivo de regresión lineal

Se construye la ecuación matemática de la línea que se graficó: Y = mX + b (donde Y son las emisiones y X es la energía).

```python
import statsmodels.api as sm

# Definimos las variables
X = df['Energía [kWh/año]'] # Variable Independiente (Predictora)
y = df['Emisiones CO2 [Ton/año]'] # Variable Dependiente (Respuesta)

# statsmodels por defecto asume que la línea pasa por el origen (0,0). 
# Debemos agregar una constante explícitamente para calcular el intercepto 'b'
X_con_constante = sm.add_constant(X)

# Construimos y ajustamos el modelo de Mínimos Cuadrados Ordinarios (OLS)
modelo = sm.OLS(y, X_con_constante)
resultados = modelo.fit()

# Reporte completo de evaluación
print(resultados.summary())
```

# Evaluación del modelo

Con los datos de la columna coef, podemos reconstruir la ecuación exacta que rige el modelo:

$$Emisiones\_CO2\_[Ton/año] = 0.0347 + (0.0008 \times Energía\_[kWh/año])$$

## Interpretación:

1. R-cuadrado = 1.000: 

El R-cuadrado indica qué porcentaje de las emisiones de CO2 que se ahorran se explica por la producción de energía. El resultado obtenido es 1.000 (100%).

En el mundo real obtener un 100% perfecto es prácticamente imposible debido a las ineficiencias, el clima o errores de los sensores. Este R-cuadrado sugiere fuertemente una relación determinista. Es decir, la columna de 'Emisiones CO2' en el DataFrame no fue medida físicamente por un sensor, sino que fue calculada matemáticamente a partir de la columna de 'Energía' multiplicándola por un factor de emisión fijo.

2. La Pendiente = 0.0008:

Esto significa que, por cada 1 kWh adicional de energía solar que se produce, las emisiones ahorradas aumentan exactamente en 0.0008 Toneladas de CO2, que equivale a 0.8 kg de CO2. En el contexto de la transición energética en Colombia, un factor de emisión de 0.8 kg CO2/kWh es bastante alto. El Sistema Interconectado Nacional (SIN), por ser mayoritariamente hidroeléctrico, suele rondar los 0.1 a 0.2 kg CO2/kWh. Ese 0.8 es más característico de generación térmica pura (como plantas de carbón o generadores diésel fuera de la red).

3. El Intercepto (const = 0.0347):
 
Físicamente representaría las emisiones cuando la producción de energía es cero. Sin embargo, la columna P>|t| para este coeficiente: es 0.133. Como es mayor a 0.05, estadísticamente este valor no es significativo, por lo que podemos asumir que el intercepto real es cero (lo cual tiene todo el sentido físico: sin energía producida, no hay ahorro de emisiones equivalentes).

El modelo confirma que los datos tienen una correlación lineal absoluta porque derivan de un cálculo directo, donde el factor de emisión aplicado fue de 0.8 kg de CO2 por cada kWh.

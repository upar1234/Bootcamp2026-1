```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('MES_1225.xlsx')
```

```python
df.head(20)
```

```python
df = df.drop(df.index[0:7], axis = 0)
df.reset_index(drop = True)
df
```

```python
df.columns
df.info()
```

```python
df.columns = df.iloc[0]
df.drop(7, inplace=True)
```

```python
df.reset_index(drop = True)
df.columns.name = None
```

```python
nulos = df.isnull()
nulos = df[nulos['Value'] == True]
nulos
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
# Extraer el número del mes (1 al 12)

df['Month'] = df['Time'].dt.month
# Extraer el año (ej. 2026)
df['Year'] = df['Time'].dt.year

colombia = df[df['Country'] == 'Colombia']
colombia.reset_index(drop = True)
```

```python
from IPython.display import FileLink

# 1. Guardas el DataFrame como CSV
colombia.to_csv('colombia.csv', index=False)

# 2. Creas un enlace de descarga interactivo
FileLink('colombia.csv')
```

```python
productos = colombia['Product'].value_counts()
productos
```

```python

# Cuenta los registros y lo convierte en una tabla organizada
resumen_tipos = colombia['Product'].value_counts().reset_index()

# Opcional: Le cambiamos el nombre a las columnas para que sea más claro
resumen_tipos.columns = ['Tipo de Proyecto', 'Cantidad de Registros']

# Mostramos el resultado
print(resumen_tipos)
```

```python
# Cuenta los registros y lo convierte en una tabla organizada
resumen_años = colombia['Year'].value_counts().reset_index()

# Opcional: Le cambiamos el nombre a las columnas para que sea más claro
resumen_años.columns = ['Año', 'Cantidad de Registros por año']

# Mostramos el resultado
print(resumen_años)
```

```python
# Cuenta los registros y lo convierte en una tabla organizada
resumen_años = colombia['Month'].value_counts().reset_index()

# Opcional: Le cambiamos el nombre a las columnas para que sea más claro
resumen_años.columns = ['Mes', 'Cantidad de Registros por mes']

# Mostramos el resultado
print(resumen_años)
```

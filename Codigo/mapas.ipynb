```python
import streamlit as st
import pandas as pd
import plotly.express as px
import requests
```

```python
url_repo = "https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/3aadedf47badbdac823b00dbe259f6bc6d9e1899/colombia.geo.json"
```

```python
@st.cache_data #Se usa la cache para evitar que se vuelva a descargar el mismo archivo cada vez que se ejecute el código
def load_json(url):
    responde = requests.get(url)
    return responde.json()
@st.cache_data
def get_data():
    dataframe = pd.read_csv('solar.csv') 
    dataframe['Departamento'] = dataframe['Departamento'].astype(str).str.upper().str.strip()
    return dataframe
```

```python

get_data()
```

```python
colombia_geojson = load_json(url_repo)
```

```python
fig = px.choropleth(
    get_data(),
    geojson=colombia_geojson,
    locations="Departamento",        # Columna de tu DataFrame
    featureidkey="properties.DPTO_CNMBR",  # Clave exacta dentro del GeoJSON
    color="Energía [kWh/año]",             # Columna de valores numéricos
    color_continuous_scale="Viridis",
    labels={'PRODUCCION': 'Generación (MWh)'},
    title="Mapa de Calor de Generación"
)
```

```python
st.write("""
# My first app
Hello *world!*
""")
```

```python
fig.update_geos(fitbounds="locations", visible=False)
st.plotly_chart(fig, use_container_width=True)

```

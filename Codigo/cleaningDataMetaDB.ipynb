3. Inconsistencias de formato (Unificar formatos de fechas, números, mayúsculas/minúsculas) y Conversión de tipos de datos (Asegurar que cada columna tenga el tipo de dato correcto)

```python
# Unificar texto a capitalizados (la primera letra de cada palabra en mayúscula)
# para las columnas de texto: Proyecto,Tipo,Departamento,Municipio
df1['Proyecto'] = df1['Proyecto'].str.capitalize()
df1['Tipo'] = df1['Tipo'].str.capitalize()
df1['Departamento'] = df1['Departamento'].str.capitalize()
df1['Municipio'] = df1['Municipio'].str.capitalize()
df1
```

```python
# Unificar formato de fechas
df1['Fecha estimada FPO'] = pd.to_datetime(df1['Fecha estimada FPO'], errors='coerce') 
df1 # Ajusta el nombre de la columna
```

```python
#Revisar los datos registrados en columnas numéricas
df1['Empleos estimados'].unique()
```

```python
#convertir a numérico, forzando errores a NaN para identificar problemas de formato
#df1["Empleos estimados"] = pd.to_numeric(df1["Empleos estimados"], errors="coerce")
try:
    df1["Empleos estimados"] = (
        df1["Empleos estimados"]
        .str.replace(',', '')
        .str.strip()
        .pipe(pd.to_numeric, errors='coerce')
    )
except Exception as e:
    pass
```

```python
#comprobar que se hizo la conversión correctamente
df1['Empleos estimados'].unique()
```

```python
#Revisar los datos registrados en columnas numéricas
df1['Usuarios'].unique()
```
